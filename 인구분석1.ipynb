```python
import csv
import matplotlib.pyplot as plt

from matplotlib import rc
plt.rc('font', family='NanumBarunGothic') 

f=open('2021_gender.csv','r',encoding='CP949')
lines=csv.reader(f)
m=[]
f=[]
for line in lines:
    if '상암동' in line[0]:
        for i in line[3:104]:
            m.append(-int(i))
        for i in line[106:]:
            f.append(int(i)) 
plt.title('상암동의 인구분포')
plt.barh(range(101),m, label='남성')
plt.barh(range(101),f, label='여성')
plt.legend()
plt.show()
```
